## 读取数据

导入数据分析需要的库，并通过Pandas的read_scv函数，将原始数据"Online Retail.csv"里的数据内容，解析为DataFrame,并赋值给original_data

```python
import pandas as pd
```

```python
original_data = pd.read_csv("Online Retail.csv")
```

```python
original_data.head()
```

## 评估数据

在这一部分，我将对在上一部分建立的 original_data 这个DataFrame所包含的数据进行评估。

评估主要从两个方面进行：结构和内容，即整齐度和干净度。数据的结构性问题指不符合“每列是一个变量，每行是一个观察值，每个单元格是一个值”这三个标准，数据的内容性问题包括存在丢失数据、重复数据、无效数据等。

### 评估数据整齐度

```python
original_data.sample(10)
```


从抽样结果来看，符合每行是一个变量，每列是一个观察者，每个单元格是一个值。具体来看每行是商品的一次交易，每列是相关的变量，因此不存在结构性问题

### 评估数据的干净程度

```python
 original_data.info()
```

从输出结果来看，数据共有541,909条观察值，而Description、CustomerID变量存在缺失值。

此外，InvoiceDate的数据类型应为日期，CustomerID的数据类型应为字符串，应当进行数据格式转换。

### 评估缺失数据

在了解到description存在缺失值后，根据条件提取带有缺失值的行

```python
original_data[original_data['Description'].isna()]
```

有1454条交易数据缺失Description变量值。

从输出结果来看，这些缺失Description的交易数据，UnitPrice都为0。为了验证猜想，我们增加筛选条件，看是否存在Description变量缺失且UnitPrice不为0的数据。

```python
original_data[(original_data['Description'].isna())&(original_data['UnitPrice'] != 0.0)]
```

筛选结果显示0条记录，说明缺失Description值的数据同时也不具备有效的UnitPrice值。

Description（产品名称）和UnitPrice（产品单价）是商品交易分析的关键变量。当这两个字段同时缺失/无效时，数据无法提供有效信息，建议删除这些记录。

-----------------
CustomerID同意存在缺失值

```python
original_data[original_data['CustomerID'].isna()]
```

CustomerID（客户编号）不是分析畅销商品的必要变量。根据输出结果显示，部分CustomerID缺失的交易数据仍然有效，因此保留这些CustomerID为空的观察值。

### 评估重复数据

InvoiceNo（发票编号）：单次交易可能包含多件商品，因此允许重复

StockCode（商品编码）：不同交易可能包含相同商品，因此允许重复

CustomerID（客户编号）：同一客户可进行多次交易，因此允许重复

### 评估不一致数据

不一致数据可能存在Country变量中，我们要查看是否存在多个值指代同一个国家的情况

```python
original_data["Country"].value_counts()
```

美国存在多种表示："USA"、"United States"

英国存在多种表示："United Kingdom"、"UK"、"U.K."

### 评估无效或错误数据

可以通过DataFrame的`describe`方法，对数值的统计信息进行快速的了解

```python
original_data.describe()
```

退货订单通常表现为负Quantity

价格补偿/调整可能产生负UnitPrice

```python
original_data[original_data['Quantity'] < 0]
```

从筛选结果来看，Quantity变量为负数的观察值，InvoiceNo似乎均以C开头，表示订单被取消。

但此猜测不一定准确，为了验证猜想，我们增加筛选条件，看是否存在Quantity变量为负数且InvoiceNo不以C开头的观察值。

```python
original_data[(original_data['Quantity'] < 0) & (original_data['InvoiceNo'].str[0] != "C")]
```

以上猜想被证实错误，因为还存在Quantity变量为负数且InvoiceNo不以C开头的观察值。

但根据输出结果，这些筛选出的观察值的UnitPrice观察值均为0，因此需要增加UnitPrice的条件进行进一步验证。

```python
original_data[(original_data['Quantity'] < 0) & (original_data['InvoiceNo'].str[0] != "C") & (original_data['UnitPrice'] != 0)]
```

根据验证结果，Quantity为负数的记录符合以下特征之一：

InvoiceNo以"C"开头（表示取消的订单）

UnitPrice为0（无效交易价格）

这些记录均不构成有效销售数据，将在清洗阶段被移除。

-----
接下来筛选UnitPrice也是负数的行

```python
original_data[(original_data['UnitPrice'] < 0)]
```

根据分析结果，UnitPrice为负数的记录属于坏账调账（非实际商品交易数据），应在数据清洗中予以删除。

## 清洗数据

根据前面评估部分得到的结论，我们需要进行的数据清理包括：

- 把 InvoiceDate 变量的数据类型转换为为日期时间
- 把 CustomerID 变量的数据类型转换为字符串
- 把 Description 变量缺失的观察值删除
- 把 Country 变量值 "USA" 替换为 "United States"
- 把 Country 变量值 "UK" 、 "U.K." 替换为 "United Kingdom"
- 把 Quantity 变量值为负数的观察值删除
- 把 UnitPrice 变量值为负数的观察值删除

为了区分清理前后的数据，我们创建 cleaned_data 作为 original_data 的副本，后续所有清理操作将在 cleaned_data 上执行：

```python
cleaned_data = original_data.copy()
cleaned_data.head()
```

把 InvoiceDate 变量的数据类型转换为为日期时间

```python
cleaned_data['InvoiceDate'] = pd.to_datetime(cleaned_data['InvoiceDate'])
cleaned_data['InvoiceDate']
```

把 CustomerID 变量的数据类型转换为字符串

```python
cleaned_data["CustomerID"] = cleaned_data["CustomerID"].astype(str)
cleaned_data["CustomerID"]
```

我们要把`CustomerID`结尾的`.0`删除

```python
cleaned_data["CustomerID"] = cleaned_data["CustomerID"].str.slice(0,-2)
cleaned_data["CustomerID"]
```

把 Description 变量缺失的观察值删除

```python
cleaned_data.dropna(subset=["Description"],inplace=True)
cleaned_data["Description"].isna().sum()
```

把 Country 变量值 "USA" 替换为 "United States"

```python
cleaned_data["Country"] = cleaned_data["Country"].replace({"USA":"United States"})
```

```python
len(cleaned_data[cleaned_data["Country"] == "USA"])
```

把 Country 变量值 "UK" 、 "U.K." 替换为 "United Kingdom"

```python
cleaned_data["Country"] = cleaned_data["Country"].replace({"UK":"United Kingdom","U.K":"United Kingdom"})
```

```python
print(len(cleaned_data[cleaned_data["Country"] == "UK"]))
print(len(cleaned_data[cleaned_data["Country"] == "U.K"]))
```

把 Quantity 变量值为负数的观察值删除

```python
cleaned_data = cleaned_data[(cleaned_data['Quantity'] >= 0)]
```

```python
len(cleaned_data[(cleaned_data['Quantity'] <= 0)])
```

把 UnitPrice 变量值为负数的观察值删除

```python
cleaned_data = cleaned_data[(cleaned_data['UnitPrice'] >= 0)]
```

```python
len(cleaned_data[(cleaned_data['UnitPrice'] < 0)])
```

```python
cleaned_data.sample(10)
```

## 保存清理后的数据

```python
cleaned_data.to_csv("Online Retail_cleaned.csv",index = False)
```
